## Main task: Combining headers

```python
#Definition=> this combines the headers and replaces the target-row header;
#Parameters=> target-row: is the main header,row_tocheck: is the number of sub-headers need to be combined
#Return=> returns the number of header rows that has been checked, it'll be usefull to drop the redundant rows.
def combineHeaders(target_row, row_tocheck, data):
    target_headers = list()
    target_headers = data.iloc[[target_row]]
    row_checked = list();
    for i in range(row_tocheck):
        next_row = target_row + i + 1;
        for col in range(data.shape[1]):
            #FirstIf=> To check wheather the next column is header row or data row, if data row than no need to go further
            if((col==1) and (target_headers.iloc[0,col]!=data.iloc[next_row,col])):
                #print('Data row:',next_row)
                return row_checked;
            if ((col!=0) and (target_headers.iloc[0,col]!=data.iloc[next_row,col])): # ignore column 0
                
                # update
                data.iloc[target_row,col] = data.iloc[target_row,col] + ' ' + data.iloc[next_row,col]
        #print('Header Rows: ',next_row)
        row_checked = row_checked + [next_row]
    return data, row_checked
```

```python
newdata = original_data.copy();#just using a new dataframe to avoid changing the main one
newdata,row_checked = combineHeaders(2,2,newdata)
newdata[:6]
```

```python
# drop redundant rows
if row_checked:
    for row in row_checked:
        newdata.drop([row], axis=0, inplace=True)

# drop one extra column
newdata.drop(columns[0], axis=1, inplace=True)
newdata[:5]
```

```python
# reset the index after dropping rows
newdata = newdata.reset_index(drop=True)
newdata[:5]
```
